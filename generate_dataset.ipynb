# Synthetic Dataset Generation

```python
import pandas as pd
import random

categories = {
    "Water Issue": [
        "Water leakage for {} days",
        "No proper water supply",
        "Pipeline burst near market",
        "Dirty water coming from taps",
        "Low water pressure in area"
    ],
    "Garbage": [
        "Garbage not collected for {} days",
        "Overflowing dustbins near road",
        "Waste dumping issue",
        "Bad smell due to garbage",
        "Garbage scattered everywhere"
    ],
    "Road Damage": [
        "Pothole causing accidents",
        "Road cracked badly after rain",
        "Uneven road surface",
        "Big pothole near school",
        "Road sinking problem"
    ],
    "Electricity": [
        "Streetlight not working",
        "Power outage for {} hours",
        "Loose electric wires hanging",
        "Frequent power cuts",
        "Electric pole sparking"
    ],
    "Drainage": [
        "Drain blockage for {} days",
        "Water logging problem",
        "Sewage overflow",
        "Drain cleaning required urgently",
        "Drain water mixed on road"
    ]
}

location_types = ["Residential", "Market", "Highway", "School Area", "Hospital Area"]

noise_phrases = [
    "please resolve soon",
    "very urgent",
    "since many days",
    "causing problems",
    "citizens suffering",
    "kindly look into this",
    "not fixed yet",
    "immediate action needed",
    "dangerous situation"
]

typo_words = {
    "water": "watr",
    "garbage": "garbge",
    "road": "raod",
    "drain": "dran",
    "electric": "eletric"
}

data = []

for _ in range(2000):

    category = random.choice(list(categories.keys()))
    text_template = random.choice(categories[category])
    days = random.randint(1, 10)

    # Fill numeric placeholders
    if "{}" in text_template:
        text = text_template.format(days)
    else:
        text = text_template

    # 25% chance: add words from another category (confusion)
    if random.random() < 0.25:
        other_category = random.choice(list(categories.keys()))
        extra_text = random.choice(categories[other_category])
        text += " and also " + extra_text

    # Add noise phrase
    if random.random() < 0.7:
        text += " " + random.choice(noise_phrases)

    # 15% chance: introduce typo
    if random.random() < 0.15:
        for word in typo_words:
            if word in text.lower():
                text = text.replace(word, typo_words[word])
                break

    location = random.choice(location_types)

    # Advanced Priority Logic
    text_lower = text.lower()

    if "accident" in text_lower or location == "Highway":
        priority = "Critical"
    elif "danger" in text_lower or "sparking" in text_lower:
        priority = "High"
    elif days > 6:
        priority = "High"
    elif days > 3:
        priority = "Medium"
    else:
        priority = "Low"

    data.append([text, category, days, location, priority])

df = pd.DataFrame(data, columns=[
    "complaint_text",
    "category",
    "days_pending",
    "location_type",
    "priority"
])

df.to_csv("civic_complaints.csv", index=False)

print("Dataset Generated Successfully!")
```
